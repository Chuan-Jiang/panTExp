```r
library(tidyverse)
```

```r
softmax <- function(x) {
  exp(x) / sum(exp(x), na.rm = TRUE)
}
```

```r
raw <- read_tsv("../HERV.raw.count.tsv") |> rename(TE = 1, Read = 2, Map = 3, AS = 4)

head(raw)
```

```r
cm <-  raw    |> pivot_wider(id_cols = TE, names_from = Read, values_from = AS,  values_fn = max) 
pi = 1/dim(cm)[1]

cm[seq(1,10),seq(1,10)]
```

```r
cmsoft <- cm |> mutate(across(-TE, softmax))
```

```r
cmsoft[seq(1,10),seq(1,10)]
```
