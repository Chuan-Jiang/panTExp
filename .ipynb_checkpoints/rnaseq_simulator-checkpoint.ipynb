### GRCH38   
each batch randomly select 1000 transcripts, the accoaiste counts starts from 1 reads - 500reads

```r
fasta_file <- "../04_simulation/GRCh38.HERVK.fa"
fasta = readDNAStringSet(fasta_file)
set.seed(4)

n <- 100

for(b in seq(1,10) ) {
    
    fasta_subset <- fasta[sample(length(fasta), n)]

    dir_name =  paste0("../04_simulation/GRCh38_Batches",b)
    cat("working on dir:", dir_name, "\n")
    if (!dir.exists(dir_name)) {
      dir.create(dir_name, recursive = TRUE)
    }

    fa.file <- paste0(dir_name,"/batch.",b,".fasta")
    writeXStringSet(fasta_subset, filepath = fa.file )
    
    random_numbers <- runif(n,min = 0, max = 1)
    # Scale values it range [30, 300]
    min_val <- 1
    max_val <- 300
    # Rescale beta values to desired range
    readspertx <- as.integer(min_val + (max_val - min_val) * random_numbers)

    ctl <- rep(1, n)
    trt <- sample(c(0.5, 1, 2), size=n,
        prob=c(0.05, 0.9, 0.05), replace=TRUE)
    fc <- cbind(ctl, trt)

    simulate_experiment(fa.file, reads_per_transcript=readspertx, seed = 1 ,readlen = 150,fraglen = 400,fragsd = 50 , 
             num_reps=c(2,2) , outdir= dir_name , strand_specific = TRUE, fold_changes =  fc  )

}
```

```r
# FASTA annotation
#fasta_file = system.file('extdata', 'chr22.fa', package='polyester')
fasta_file <- "../04_simulation//HG002.mat.HERVK.fa"
fasta = readDNAStringSet(fasta_file)

set.seed(123)
n <- length(fasta)
alpha <- 1  # Shape parameter α (controls skewness)
beta <- 5   # Shape parameter β (controls skewness)
# Generate beta-distributed values
random_numbers <- rbeta(n, alpha, beta)
# Scale values it range [30, 300]
min_val <- 1
max_val <- 300
# Rescale beta values to desired range
readspertx <- as.integer(min_val + (max_val - min_val) * random_numbers)

fold_changes = rep(1 , n)
fsum <- sum(readspertx > 30)
fold_changes[readspertx > 30] = sample(c(0.5, 1, 2), size=fsum, prob=c(0.1, 0.8, 0.1), replace=TRUE)
#fold_changes = sample(c(0.5, 1, 2), size=n, prob=c(0.05, 0.9, 0.05), replace=TRUE)
#readspertx > 

simulate_experiment(fasta_file, reads_per_transcript=readspertx, seed = 1,readlen = 150,fraglen = 400,fragsd = 50,
    num_reps=c(5,5), outdir='../04_simulation/simulated_reads1', fold_changes = fold_changes, strand_specific = TRUE) 
```
